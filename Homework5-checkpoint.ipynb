    요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> c = c + 5  
        >>> d = d - 5  
        >>> print(c)  
        [Count (step: 1)] 17

    요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> print(c) > 10  
        True  
        >>> print(d) > 10  
        False  
        >>> print(c) < 10  
        False  
        >>> print(d) < 10  
        True  
        >>> print(c) == 17  
        True  
        >>> print(d) != 9  
        False

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return "[Count (step: {})] {} ".format(self.step, self.count)
    
    def incr(self): 
        self.count = self.count + self.step
        
    def __call__(self): 
        return self.incr()
        
    def __add__(self, n):
        return Counter(self.count + n)
        
    def __sub__(self, n):
        return Counter(self.count - n)    
    
    def __eq__(self, n):
        return self.count == n
    
    def __lt__(self, n):
        if self.count > n:
            return True
        return False
    
    def __ne__(self, n):
        return self.count != n
```

4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

        MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

init() 함수는 생성자이다. 이 함수는 인스턴스 객체가 생성될 때 자동적으로 호출된다. MySet([1, 2, 3, 4]) 클래스 객체인 s가 생성되면서 init() 함수를 호출한다. 이때 init()의 두 개의 인자 중 self는 인스턴스 객체 s이고 ㅣ은 list를 의미한다. 객체 s에 append() 함수를 통해 list를 복사하고 elimicate_duplicate() 함수를 호출한다. elimicate_duplicate() 함수는 장식자  @staticmethod를 사용하여 인스턴스 객체 없이 호출할 수 있다. 이 함수는 중복되는 값을 지우는 기능으로 인자 self, 즉 s를 받고 (s == l) 빈 list s에 중복되지 않은 ㅣ값을 저장한 뒤 최기화된 l에 옮긴다. str()은 print(s) 형태일 때 호출이된다. 객체 s를 print할 때 __str__()함수가 호출되고 형식에 맞추어 출력된다.

5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

    정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.  
    요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    >>> u = s | t  
    >>> print(u)  
    MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}

    요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

    >>> u = s & t  
    >>> print(u)  
    MySet: {2, 3}

    요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

    >>> s = MySet([1, 2, 3])  
    >>> t = MySet([3, 4, 5])  
    >>> u = s - t  
    >>> print(u)  
    MySet: {1, 2}

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    def __or__(self, other):
        s = []
        for e in self:
            s.append(e)
            
        for e in other:
            if e not in s:
                s.append(e)
                
        return s
                
    def __and__(self, other):
        s = [] 
        for e in other:
            if e in self:
                s.append(e)
        return s
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    u = s | t
    print(u)
    u = s & t
    print(u)
```

```python
class SearchEngine:
    def __init__(self, *url):
        self.url = list(url)
        self.dic_count_lst = []
        self.read_data = []

    def __str__(self):
        print("url: {}".format(self.url))

    def  addUrl(self, url): 
        if url not in self.url:
            self.url.append(url)

    def removeUrl(self, url):
        if url in self.url:
            self.url.remove(url)

    def listUrls(self):
        for url in self.url:
            print(url)

    def getWordsFrequency(self):
        n = 0
        for url in self.url:
            n = n + 1
            if url is None:
                print("Error!")
                break
            req = requests.get(url)
            source = req.text
            
            for i in range(len(source)):
                if source.find('\n') >= 0:
                    none_line = source.replace(source[source.find('\n')], '')

            length = len(source)
            text_line = self.organize(none_line, length) 
            text_line = text_line.split() 
                
            for text in text_line:
                if text == "=":
                    pos = text_line.index(text)
                    del text_line[pos -1 : pos + 2]

            text_line = self.punctuation_character(text_line)
            text_line = self.delete_bool(text_line)
            
            dic_count= self.frequency(text_line)
            self.dic_count_lst.append(dic_count)
            if dic_count is None:
                print("빈 리스트\n")
            else:
                print("{}번째 WordsFrequency ->".format(n))
                print(dic_count)
            print('\n')
                
    def organize(self, none_line, length):#문자열 none_line
        index = 0
        while True:#문자열 첫 인덱스부터 check
            if none_line[index] in '</':#값이 </로 시작하면 
                closing_index = self.meet_closing(none_line, index, length)
                text = none_line[index:closing_index + 1]
                if text == "</head>" or text == "</div>" or text == "</script>" or text == "<script>":
                    closing_index = self.meet_same(none_line, closing_index, length)
                    none_line = none_line.replace(none_line[index:closing_index + 1], "")
                else:
                    none_line = none_line.replace(none_line[index:closing_index + 1], "")#</부터 >인덱스까지 문자열 제거
                length = length - (closing_index - index + 1)#전체 길이 다시 초기화

            else:#값이 </로 시작하지 않으면 
                index += 1#계속 문자열 탐색

            if self.errorlength(none_line, index) == True:#while 문 break
                break

        return none_line#문자열 return

    def meet_same(self, none_line, closing_index, length):
        for i in range(closing_index + 1, length):
            if none_line[i] in '<':
                reclosing_index = self.meet_closing(none_line, i, length)
                text = none_line[i : reclosing_index + 1]
                if text == "</div>" or text == "</html>":
                    return reclosing_index

    def meet_closing(self, none_line, index, length):#클로징 문자열 > 인덱스 찾기
        for i in range(index , length):
            if none_line[i] in '>':
                return i

    def errorlength(self, none_line, index):#while문 제어
        return len(none_line) == index

    
    def punctuation_character(self, text_line):#count 
        text_line = ' '.join(text_line)
        for text in text_line: 
            ot = ord(text)
            if (ot > 32 and ot < 48) or (ot > 57 and ot < 65) or ( ot > 90 and ot < 97) or (ot > 122 and ot < 128):#구두문자
                text_line = text_line.replace(text, "")
        text_line = text_line.split() 
        return text_line

    def frequency(self, text_line):
        dic_count = {}
        for text in text_line:
            if text in dic_count:
                dic_count[text] += 1
            else:
                dic_count[text] = 1
        return dic_count
    
    def getMaxFreqencyWords(self): 
        n = 0
        if self.dic_count_lst is None:
            return None
        for dic_lst in self.dic_count_lst:
            n = n + 1
            read_data = dict(sorted(dic_lst.items(), key = lambda x:x[1], reverse = True))
            self.read_data.append(read_data)
            print("{}번째 MaxFreqencyWords ->".format(n))

            cnt = 0
            for key in read_data.keys():
                cnt += 1
                if cnt == 4:
                    break
               
                print("{} : {}".format(key, read_data[key]), sep = ', ')
            print("\n")
        
    def delete_bool(self, text_line): 
        stop_words = tuple(stopwords.words('english')) 

        result = []
        for w in text_line: 
            if w not in stop_words: 
                result.append(w) 

        with open("Hanbool.txt", "r", encoding='UTF-8') as han:
            stop_words = tuple(han.read().split('\n'))
            result = [word for word in result if not word in stop_words]

        return result 
    
    def searchUrlByWord(self, word):
        n = 0
        word = word.split(" ")
        inp1 = word
        dict_similar = {}
        for read_data in self.read_data:
            n = n + 1
            inp2 = set(read_data)
            similarity01 = self.JaccardSimilarity(inp1, inp2)
            print("{}번째 similarity ->".format(n))
            print(similarity01)
            index = self.read_data.index(read_data)
            
            if self.url[index] not in dict_similar:
                dict_similar[self.url[index]] = similarity01
            print('\n')
                
        dict_similar = sorted(dict_similar.items(), key = lambda x: x[1], reverse = True)
        
        for i, val in enumerate(dict_similar, 1):
            print("{}. {}".format(i, val[0])) 
        
    def JaccardSimilarity(self,inp1, inp2):
        mom = set(inp1).union(set(inp2))
        son = set(inp1).intersection(set(inp2))

        return len(son)/len(mom)
```

```python
class  SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self, *url):
        super().__init__(*url)
        self.dic_count = []
        
    def getWordsFrequency(self):
        n = 0
        for url in self.url:
            n = n + 1
            req = requests.get(url)
            source = req.text
            
            for i in range(len(source)):
                if source.find('\n') >= 0:
                    none_line = source.replace(source[source.find('\n')], '')

            length = len(source)
            text_line = self.organize(source, length) 
            text_line = text_line.split() 

            for text in text_line:
                if text == "=":
                    pos = text_line.index(text)
                    del text_line[pos -1 : pos + 2]
                    
            text_line = self.punctuation_character(text_line)
            text_line = self.delete_bool(text_line)
            
            dic_count= self.frequency(text_line)
            dic_count = sorted(dic_count.items(), key = lambda x: x[1], reverse = True)
            self.dic_count.append(dic_count)
            
            print("{}번째 WordsFrequency ->".format(n))
            print(dic_count)
                
    def organize(self, none_line, length):#문자열 none_line
        index = 0           
        while True:#문자열 첫 인덱스부터 check
            if none_line[index] in '</':#값이 </로 시작하면 
                closing_index = self.meet_closing(none_line, index, length)
                text = none_line[index:closing_index + 1]
                if text == "</head>" or text == "</div>" or text == "</script>" or text == "<script>":
                    closing_index = self.meet_same(none_line, closing_index, length)
                    none_line = none_line.replace(none_line[index:closing_index + 1], "")
                else:
                    none_line = none_line.replace(none_line[index:closing_index + 1], "")#</부터 >인덱스까지 문자열 제거
                length = length - (closing_index - index + 1)#전체 길이 다시 초기화

            else:#값이 </로 시작하지 않으면 
                index += 1#계속 문자열 탐색

            if self.errorlength(none_line, index) == True:#while 문 break
                break

        return none_line#문자열 return

    def meet_same(self, none_line, closing_index, length):
        for i in range(closing_index + 1, length):
            if none_line[i] in '<':
                reclosing_index = self.meet_closing(none_line, i, length)
                text = none_line[i : reclosing_index + 1]
                if text == "</div>" or text == "</html>":
                    return reclosing_index

    def meet_closing(self, none_line, index, length):#클로징 문자열 > 인덱스 찾기
        for i in range(index , length):
            if none_line[i] in '>':
                return i

    def errorlength(self, none_line, index):#while문 제어
        return len(none_line) == index

    
    def punctuation_character(self, text_line):#count 
        text_line = ' '.join(text_line)
        for text in text_line: 
            ot = ord(text)
            if (ot > 32 and ot < 48) or (ot > 57 and ot < 65) or ( ot > 90 and ot < 97) or (ot > 122 and ot < 128):#구두문자
                text_line = text_line.replace(text, "")
        text_line = text_line.split() 
        return text_line

    def frequency(self, text_line):
        dic_count = {}
        for text in text_line:
            if text in dic_count:
                dic_count[text] += 1
            else:
                dic_count[text] = 1
        return dic_count
    
    def delete_bool(self, text_line): 
        stop_words = tuple(stopwords.words('english')) 

        result = []
        for w in text_line: 
            if w not in stop_words: 
                result.append(w) 

        with open("Hanbool.txt", "r", encoding='UTF-8') as han:
            stop_words = tuple(han.read().split('\n'))
            result = [word for word in result if not word in stop_words]

        return result 
    
    def __iter__(self):
        for dic_count in self.dic_count:
            self.dcount = dic_count
        self.index = 0
        return self
    
    def __next__(self):  
            if self.index < len(self.dcount):
                self.index += 1
                return "({}, {})".format(self.dcount[self.index - 1][0], self.dcount[self.index - 1][1])
            else : 
                raise StopIteration     # 읽을 수 없으면 예외 발생   
```
