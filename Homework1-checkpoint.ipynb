```python
def reverse(n):
    for i in range(len(n)):
        new_n[i] = n[len(n) - ( i + 1 )]
        i += 1
    return new_n
```

8. 사용자로 부터 정수를 입력받아서 1부터 그 사이에 존재하는 소수 (Prime number)를 출력하는 파이썬 프로그램을 작성하시오

```python
n = int(input("정수를 입력하시오: "))
count = 0
```

```python
def Prime_number(n, count):
    for prime_num in range(1,n):
        if(prime_num == 1):#1은 소수에서 제외
            prime_num += 1
            continue
        else:#1을 제외한 나머지 수에서 소수 판별
            for i in range(1,prime_num + 1):#1부터 prime_num까지 나누었을 때 
                if(prime_num % i == 0):#나머지가 0이 되는 count가 2개면
                    count += 1#즉, 1과 나 자신만 있을 경우 소수가 된다.
                i += 1
            if(count == 2):
                print(prime_num, end=" ")
            prime_num += 1
            count = 0
```

```python
Prime_number(n, count)
```

input() 함수는 기본적으로 str이다. 따라서 경우에 따라 type변환을 해준다. 소수를 판별하기 위해서는 이중 for문이 필요하다.  
소수는 1과 자신만을 약수로 가지고 있는 수이다.  
따라서 소수 판별 수(prime_num)가 1부터 prime_num까지 나누었을 때 나머지가 0이 되는 개수가 2여야 한다. 즉, count를 증가시켜 나머지가 0이 되는 개수를 판별하면 된다.

9. Kevin과 Stuart는 게임을 한다. 규칙은 아래와 같다.

* 두 플레이어에게 같은 string, S, 가 주어진다.
* 두 플레이어는 string S의 문자를 사용하여 substring을 만들어야 한다(중복 가능).
* Stuart는 자음으로 시작하는 단어를 만들어야 한다.
* Kevin은 모음으로 시작하는 단어를 만들어야 한다.
* 게임은 두 플레이어가 만들 수 있는 모든 단어를 만들었을 때 끝난다.
* 점수 계산 : substring 1개당 +1 점을 획득한다. (아래 그림 참고)

* 사용자로 부터 대문자로 이루어진 string S를 입력받고, 승자와 그의 점수를 출력하는 프로그램을 작성하시오.

    [실행 예]

      string S를 입력하세요: BANANA
      승자: Stuart 점수: 12

```python
s = input('stirng S를 입력하시오: ')
KEVIN_SCORES = 0
STUART_SCORES = 0
```

```python
def scores(s, KEVIN_SCORES, STUART_SCORES):#각 플레이어의 점수 반환
    for i in range(len(s)):
        if s[i] in "A E I O U":#모음
            KEVIN_SCORES += count_vowels_words(s[i], s)#count_vowels_words() 함수 호출
            s = '!' * (i + 1) + s[i + 1:]
        else:#자음
            STUART_SCORES += count_consonants_words(s[i], s)#count_consonants_words() 함수 호출
            s = "!" * (i + 1) + s[i + 1:]
        i += 1
    return (KEVIN_SCORES, STUART_SCORES)
```

```python
def count_vowels_words(val, s):#모음으로 만드는 string 개수 반환
    SCORES = 0
    index = s.find(val)
    for j in range(index, len(s)):
       SCORES += 1
    return SCORES
```

```python
def count_consonants_words(val, s):#자음으로 만드는 string 개수 반환
    SCORES = 0
    index = s.find(val)
    for j in range(index, len(s)):
        SCORES += 1
    return SCORES
```

```python
KEVIN_SCORES, STUART_SCORES = scores(s, KEVIN_SCORES, STUART_SCORES)#scores() 호출

if(KEVIN_SCORES > STUART_SCORES):
    print('승자: KEVIN, 점수:',KEVIN_SCORES)
elif(KEVIN_SCORES < STUART_SCORES):
    print('승자: STUART, 점수:', STUART_SCORES)
else:
    print('비겼다, 점수:', KEVIN_SCORES)
```

우선 모음으로 시작하는 단어 개수를 확인하는 함수, 자음으로 시작하는 단어 개수를 확인하는 함수, 자음과 모음을 판별해 각각에 맞는 함수를 호출하는 함수로 나눌 수 있다. 모음 판별은 for문을 통해 s[i]를 반복하여 "A E I O U"가 있는지 확인한다.  
모음이 있으면 count_vowels_words()함수를 호출하고 없다면 count_consonants_words()함수를 호출한다.  
인자는 s[i]와 string S를 함수에 넘긴다. s.find()함수를 통해 s[i]값의 index를 저장하고 index부터 마지막까지 SCORES를 하나씩 증가한다. 이때 동일한 s[i]값은 index가 같기에 분별하기 위해서 판별한 s[i]의 값을 문자 '!'로 바꾼다.  
- 단어 개수 세는 방법: string S의 모음과 자음을 분리하고, 선택한 단어부터 배열 끝까지 SCORES += 1하면 된다.

10. 알파벳 소문자만으로 이루어진 string, S, 가 주어지고 아래 2가지 연산을 수행할 수 있다.
* 추가 : string S의 끝에 알파벳 소문자 하나를 추가한다.
* 삭제 : string S의 마지막 문자 하나를 삭제한다. empty string에서 삭제 연산을 수행한 결과는 empty string이다.  

두 개의 string, s와 t, 그리고 정수 k가 주어질 때, 정확히 k번의 연산을 통해 s를 t로 바꿀 수 있는지 판단하는 함수를 작성하시오.

```python
index = 0
```

```python
s = input('string s를 입력하시오: ')
t = input('string t를 입력하시오: ')
k = int(input('k를 입력하시오: '))
```

```python
def delete(s, t, k, index):
    for index in range(len(s)):
        if not s[index] == t[index]:#index = 6
            re_s = s.rstrip(s[index:])#len(s) = 11
            break
        index += 1    
    k = k - (len(s) - index)#k = 4
    return k, re_s, index
```

```python
def add(s, t, k, index):
    s = s + t[index:]
    k = k - len(t[index:])
    return s, k
```

```python
def appendDelete(s, t, k, index):
    if len(s) == len(t):
        if k % 2 == 0:
            return 'Yes'
        else:
            return 'No'
    else:
        k, s, index = delete(s, t, k, index)
        if len(s) == len(t):
            if k % 2 == 0:
                return 'Yes'
            else:
                return 'No'
        s, k = add(s, t, k, index)
        if not k == 0 :
            return 'No'
    return 'Yes' 
```

```python
print('s = {0}      t = {1}     k = {2}'.format(s, t, k), end = " => ")
print(appendDelete(s, t, k, index))
```

삭제와 추가 함수, 결과를 판별하는 함수(appendDelete)가 필요하다. 우선 s와 t의 길이가 같으면 k가 짝수여야 return값이 'Yes'가 된다.  
s와 t의 길이가 다르면 먼저 delete()함수를 호출한 뒤 k 값에 따라서 add()함수를 호출한다.  
for 문을 통해 s[index]가 t[index]와 다른 index를 찾은 뒤, index부터 s의 오른쪽 문자열을 지운다. 이때 s.rstrip()함수가 사용된다.  
바뀐 s와 t의 길이가 같으면 위와 같이 k가 짝수여야 return값이 'Yes'가 된다. s와 t의 길이가 다르면 add()함수를 호출하여 t 문자열을 s에 추가한다.
